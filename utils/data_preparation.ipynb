```python
import pandas as pd
```

```python
df = pd.read_hdf('data/h5/combined_data.h5')
df
```

```python
# Total unique objects
Total_unique_objects = df['OBJ_NUMERIS'].nunique()
Total_unique_objects
```

```python
# Total hours between dates
unique_hours = df['PL_T'].nunique()
unique_hours
```

```python
# Clean NaN values for P+ as the P- doesn't affect electricity usage
df = df[df['P+'].notna()]

# Drop all missing dates
id_counts = df.groupby('OBJ_NUMERIS').size()
complete_ids = id_counts[id_counts == 17520].index
df = df[df['OBJ_NUMERIS'].isin(complete_ids)]
df
```

```python
# Objects with full data
obj_unique_count = df['OBJ_NUMERIS'].nunique()
obj_unique_count
```

```python
#Check missing values
missing_values = df.isna().sum()
missing_values
```

```python
#Fill missing values with zeros
df["P-"] = df[ "P-"].fillna(0)
missing_values = df.isna().sum()
missing_values
```

```python
# Save data to a new h5 file
df.to_hdf('data/h5/cleaned_data.h5', key='data', mode='w')
```
